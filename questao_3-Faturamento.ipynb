```python
import pandas as pd
import json
```

```python
with open('./dados.json') as arquivo_json:
    objeto_json = json.load(arquivo_json)

objeto_json
```

```python
df = pd.DataFrame(objeto_json)

df
```

```python
exclude_zero = df['valor'] > 0
new_df = df[exclude_zero]

new_df
```

```python
menor_valor = new_df['valor'].min()

menor_valor
```

```python
maior_valor = new_df['valor'].max()

maior_valor
```

```python
acima_da_media = new_df['valor'] > (new_df['valor'].mean())
dias_acima_da_media = new_df[acima_da_media]['valor'].count()

dias_acima_da_media
```
